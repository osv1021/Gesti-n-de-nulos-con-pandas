```python
import numpy as np
import pandas as pd
```

```python
#Para poder ver todas las filas y todas las columnas del DataFrame
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df=pd.read_csv("online_sales_nulos.csv", parse_dates=['Invoice_Date','Date','Hour'])
df.head()
```

```python
#Información general del DataFrame
df.info()
```

```python
#Número de nulos por columna
df.isnull().sum() 
```

```python
nulos=(df.isnull().sum() / df.shape[0]) * 100
```

```python
#Creo un DataFrame con el % de nulos por columna
df_nulos=pd.DataFrame(nulos,columns=['Nulos_%'])
df_nulos
```

```python
#Me quedo solo con los que tienen NaN
df_nulos=df_nulos[df_nulos['Nulos_%']>0].sort_values(by='Nulos_%',ascending=False)
df_nulos
```

## Imputación columnas categóricas

```python
df.head()
```

```python
#Tenemos un 6.9% de Nan. Imputamos con 'Unknow'
df['Warehouse_Location'].fillna('Unknow', inplace=True)
```

```python
#Tenemos un 4.85% de Nan. Imputamos con 'Free
df['Shipping_Cost_Category'].fillna('Free', inplace=True)
```

```python
#Han desaparecido los Nan de las columnas categóricas
df.isnull().sum()
```

## Imputación columnas numéricas

```python
df.head()
```

```python
#Convertimos 'Customer_Id' a object e imputamos por desconocido
df['Customer_Id']=df['Customer_Id'].astype('object')
df['Customer_Id'].fillna('Unknow', inplace=True)
```

```python
#Comprobamos
df['Customer_Id'].head()
```

```python
#Imputamos la columna 'Shipping_Cost' con un 0
df['Shipping_Cost'].fillna(0, inplace=True)
```

```python
#Observo la media, mediana y desviación estandar de 3 columnas
df[['Quantity','Unit_Price','Discount']].describe().T[['mean','50%','std']]
```

```python
#Decido imputar con la mediana
df['Quantity']=df['Quantity'].fillna(df['Quantity'].median())
df['Discount']=df['Discount'].fillna(df['Discount'].median())
```

```python
#Creo un DataFrame con la columna 'Unit_Price
df_Unit_Price=df['Unit_Price'].reset_index()
df_Unit_Price.head()
```

```python
#Calculo la media y mediana
media_Unit_Price=df['Unit_Price'].mean()
mediana_Unit_Price=df['Unit_Price'].median()
```

```python
#Creo dos columnas nuevas 
df_Unit_Price['media_Unit_Price']=df_Unit_Price['Unit_Price'].fillna(media_Unit_Price)
df_Unit_Price['mediana_Unit_Price']=df_Unit_Price['Unit_Price'].fillna(mediana_Unit_Price)
df_Unit_Price.head()
```

```python
#Muestro desviación estandar de las dos columnas creadas
print(df_Unit_Price['media_Unit_Price'].describe()['std'])
print(df_Unit_Price['mediana_Unit_Price'].describe()['std'])
```

```python
#Imputo 'Unit_Price' con la media
df['Unit_Price']=df['Unit_Price'].fillna(media_Unit_Price)
```

```python
#Reviso que no hay Nan en el DataFrame
df.isnull().sum()
```
